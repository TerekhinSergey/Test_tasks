Для проверки корректной отработки запроса сгенерируем необходимые данные для БД.

```python
import pandas as pd
import numpy as np
```

```python
accounts = pd.DataFrame(np.random.randint(100, 200, size=(100, 1)), columns=['acc_num']) 
accounts['rest_rub'] = np.random.randint(100, 100000, size=(100, 1))
accounts['report_date'] = np.random.choice(pd.date_range('2016-01-01', '2016-12-31'), len(accounts))
accounts
```

```python
cards = pd.DataFrame(np.random.randint(200, 300, size=(100, 1)), columns=['hash_id']) 
cards['tarif_id'] = np.random.randint(1, 4, size=(100, 1))
cards['date_issued'] = np.random.choice(pd.date_range('2016-01-01', '2016-12-31'), len(cards))
cards['date_activated'] = np.random.choice(pd.date_range('2016-01-01', '2016-12-31'), len(cards))
cards['acc_num'] = np.random.randint(100, 200, size=(100, 1))
cards['client_id'] = np.random.randint(1, 100, size=(100, 1))
cards
```

```python
catalog = (pd.DataFrame({'tarif_id': [1, 2, 3, 4],
                         'tarif_name': ['амурский тигр', 'неамурский тигр',
                                        'дальневосточный леопард', 'дальневосточный леопард']}))
catalog
```

```python
black_list = pd.DataFrame(np.random.randint(1, 100, size=(20, 1)), columns=['client_id'])
black_list
```

```python
card_status = pd.DataFrame(np.random.randint(200, 300, size=(100, 1)), columns=['hash_id']) 
card_status['date_status'] = np.random.choice(pd.date_range('2016-01-01', '2016-12-31'), len(card_status))
card_status['names'] = np.resize(['open','closed',], len(card_status))
card_status
```

```python
tel_catalog = pd.DataFrame(np.random.randint(1, 100, size=(100, 1)), columns=['client_id']) 
tel_catalog['tel_num'] = np.random.randint(89000000000, 89999999999, size=(100, 1), dtype=np.int64)
tel_catalog
```

```python
df_list = [accounts, cards, catalog, black_list, card_status, tel_catalog]
names = ['accounts.xlsx', 'cards.xlsx', 'catalog.xlsx', 'black_list.xlsx', 'card_status.xlsx', 'tel_catalog.xlsx']
```

```python
writer=pd.ExcelWriter(r"C:\\Users\\Sergey\\Documents\\Python Scripts\\generated_data.xlsx")
for i, A in enumerate(df_list):
    A.to_excel(writer, sheet_name="{0}".format(names[i]), index=False)

writer.save()
```
